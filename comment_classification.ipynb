```python
vectorizer = TextVectorization(max_tokens=max_tokens,
                              output_sequence_length=1800,
                               output_mode="int" )

```

Now I think it is time to split our dataset and preprocess our dataset

```python
data_set = tf.data.Dataset.from_tensor_slices((vectorized_text,Y))
data_set = data_set.cache()
data_set = data_set.shuffle(160000)
data_set = data_set.batch(16)
data_set = data_set.prefetch(8)
```

```python
train_data = data_set.take(int(len(data_set)*0.7))
val_data = data_set.skip(int(len(data_set)*0.7)).take(int(len(data_set)*0.2))
test_data = data_set.skip(int(len(data_set)*0.9)).take(int(len(data_set)*0.1))
```

Now it is time to create our model

```python
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Embedding(max_tokens,32))
model1.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32,activation='tanh')))
model1.add(tf.keras.layers.Dense(128,activation=tf.keras.activations.relu))
model1.add(tf.keras.layers.Dense(256,activation=tf.keras.activations.relu))
model1.add(tf.keras.layers.Dense(128,activation=tf.keras.activations.relu))
model1.add(tf.keras.layers.Dense(6,activation=tf.keras.activations.sigmoid))
```

```python
model1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
               metrics=[tf.keras.metrics.Accuracy()])
```

```python
model1.summary()
```

```python
model1.fit(train_data,
           epochs=1,
           validation_data=val_data)
```

```python
test_results = model1.predict(test_data)
```

now it is time to evaluate our model

```python
from tensorflow.keras.metrics import CategoricalAccuracy
acc = CategoricalAccuracy()
for batch in test_data.as_numpy_iterator():
  x_true,y_true = batch
  y_hat = model1.predict(x_true)

  acc.update_state(y_true,y_hat)

```

```python
print(acc.result().numpy())
```

the accuracy is hight because most are zero
we should measure recall and presicion

```python
from tensorflow.keras.metrics import Recall,Precision
```

```python
prec = Precision()
rec = Recall()

for batch in test_data.as_numpy_iterator():
  x_true,y_true = batch

  y_hat = model1.predict(x_true)
  prec.update_state(y_true,y_hat)
  rec.update_state(y_true,y_hat)

print(f"recall : {rec.result().numpy()}")
print(f"precision: {prec.result().numpy()}")
```
